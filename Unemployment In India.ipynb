### Context

The story behind this datasets is how lock-down affects employment opportunities and how the unemployment rate increases during the Covid-19.

This dataset contains the unemployment rate of all the states in India.

Key Columns in the Data
1. Region: Represents the states or territories in India.
2. Date: Indicates the time period for the unemployment data.
3. Frequency: Defines the interval (e.g., Monthly).
4. Estimated Unemployment Rate (%): The percentage of the labor force that is unemployed.
5. Estimated Employed: The total number of employed individuals.
6. Estimated Labour Participation Rate (%): The percentage of the working-age population that is part of the labor force.


##### Problem Description
The metric used to gauge unemployment is the unemployment rate, representing the proportion of unemployed individuals relative to the total labor force. The COVID-19 pandemic has led to a significant surge in the unemployment rate. This analysis aims to explore the socio-economic repercussions of the pandemic on India's workforce and labor market.

The dataset facilitates an understanding of the variations in unemployment trends among different states in India during the COVID-19 crisis. It provides valuable information on the effects on the unemployment rate, employment statistics, and labor participation rates in diverse regions of the country.

```python
area_counts = df['Area'].value_counts()

# Plot the pie chart
plt.figure(figsize=(5,5))
plt.pie(area_counts, labels=area_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Areas (Urban vs. Rural)')
plt.show()
```

```python
region_counts = df['Region'].value_counts(ascending=True)

# Sort the DataFrame by Region counts
sorted_regions = region_counts.index

# Create the count plot with sorted categories
sns.countplot(x='Region', data=df, order=sorted_regions)

# Rotate the x-axis labels by 90 degrees
plt.xticks(rotation=90)

# Display the plot
plt.title('Count of Regions (Ascending Order)')
plt.show()
```

```python
df.head()
```

```python
sns.barplot(x = 'Estimated Unemployment Rate (%)', y= 'Area', data=df)
plt.title("Unemployment Rate in Area")
plt.show()
```

```python
df.drop(['Frequency'], axis=1, inplace=True) # removing the row Frequency from table
```

```python
# Ensure the time column is in datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Group by date (if needed) and calculate the mean or sum
time_data = df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(time_data.index, time_data.values, marker='o')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Estimated Unemployment Rate (%)')
plt.grid()
plt.show()
```

```python
# Group by date and calculate the sum
time_data = df.groupby('Date')['Estimated Unemployment Rate (%)'].mean()

# Create an area plot
plt.figure(figsize=(10, 6))
plt.fill_between(time_data.index, time_data.values, alpha=0.5, color='skyblue')
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate')
plt.grid()
plt.show()
```

```python
df2 = df.groupby('Region')[['Estimated Unemployment Rate (%)']].mean().sort_values(by='Estimated Unemployment Rate (%)',ascending= False)
df2.plot(kind='bar')             #used to create a bar plot of the data in a DataFrame
plt.title('Estimated unemployment percentage per region')
plt.ylabel('Unemployment percentage')
plt.xticks(rotation= 90)
plt.figtext(x= 0.19, y= 0.7, s= 'Tripura has the highest\nunemloyment percentage\nof all regions')
plt.show()
```

The average employed per region will tell us if there a significant difference in employees among states as well as the regions with the highest and lowest number of employees.

```python
df2 = df.groupby('Region')[['Estimated Employed']].mean().sort_values(by='Estimated Employed',ascending= False)
df2.plot(kind='bar')
plt.title('Average millions employed per region')
plt.ylabel('Millions employed')
plt.figtext(x=0.15, y=0.75, s='While the numbers of employees have\na big variance among regions,\nUttar Pradesh has the most employees')
plt.show()
```
